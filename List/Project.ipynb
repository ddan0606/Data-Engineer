### Filtering Data

```python
df.filter(df['gender'] == 'Male').limit(10).toPandas()
```

```python
 df.filter(df['age'] == 'Vivo').select('age', 'ratings', 'mobile').limit(10).toPandas()
```

```python
df.filter(df['mobile'] == 'Vivo').filter(df['experience'] > 10).limit(10).toPandas()
```
